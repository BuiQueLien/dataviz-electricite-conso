# Importer la bibliotheque `pandas`

```python
import pandas as pd
```

# Importer le jeu de donnees

```python
df = pd.read_csv('Individual_Household_Electric Power_Consumption.csv')
df
```

# Analyser le jeu de donnees

```python
print(df.shape)
print(df.dtypes)
```

# Convertir la colonne `Date` de `String` a `Datetime`

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

# Ajouter la colonne `year` au dataframe

```python
df['year'] = pd.DatetimeIndex(df['Date']).year
df
```

# Calculer les consommations moyennes par annee

```python
result = df.groupby('year').mean()
result
```

# Reset index et rajouter la colonne `year` au jeu de donnees

```python
result.reset_index(inplace=True)
result
```

# Exporter le resultat au ficher .csv

```python
result.to_csv('avg-consum-by-year.csv', index=False)
```
